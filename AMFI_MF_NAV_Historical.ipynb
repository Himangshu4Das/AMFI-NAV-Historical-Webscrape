```python
import requests
from bs4 import BeautifulSoup
from pprint import pprint

# AMFI NAV data download page
url = 'https://www.amfiindia.com/nav-history-download'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

## Get list of Mutual Funds from dropdown

```python
MF_index = {}
for op in soup.find_all('select', {'class' : 'select', 'id' : 'NavDownMFName'})[0]:
    str_op = str(op)[15:].split('">')
    if len(str_op) > 1 and str_op[0] != '':
        MF_Value = str_op[0] 
        MF_Name = str_op[1].split('<')[0]
        MF_index.update({MF_Name:MF_Value})

pprint(MF_index)
```

## Get list of mutual fund types

```python
type_index = {}
for op in soup.find_all('select', {'class' : 'select', 'id' : 'NavDownType'})[0]:
    str_op = str(op)[15:].split('">')
    if len(str_op) > 1 and str_op[0] != '':
        NDT_Value = str_op[0] 
        NDT_Name = str_op[1].split('<')[0]
        type_index.update({NDT_Name:NDT_Value})
pprint(type_index)
```
